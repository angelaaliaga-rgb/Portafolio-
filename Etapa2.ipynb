## Etapa 1: Análisis descriptivo e imputación de datos

El objetivo de esta etapa es comprender la estructura del dataset del Banco Mundial, evaluar la calidad de los datos disponibles y preparar la información para las etapas posteriores de reducción de dimensionalidad y modelación.

Se realiza una revisión general del conjunto de datos, identificando el número de países, variables y observaciones, así como la presencia de valores faltantes y outliers relevantes.


````python
# Número total de observaciones
df_wb_raw.shape[0]
#El código tiene un total de 198 observaciones, correspondiente a paises individuales.
````

````python
# Porcentaje de valores faltantes por variable
na_pct = df_wb_raw.isna().mean().sort_values(ascending=False) * 100

na_table = (
    na_pct
    .reset_index()
    .rename(columns={'index': 'Variable', 0: 'Porcentaje_NA'})
)

na_table.head(15)
````

````python
# Regla del 15%
vars_eliminar = na_table.loc[na_table['Porcentaje_NA'] >= 15, 'Variable'].tolist()
vars_imputar = na_table.loc[
    (na_table['Porcentaje_NA'] > 0) & (na_table['Porcentaje_NA'] < 15),
    'Variable'
].tolist()

print("Variables a eliminar (>=15% NA):", len(vars_eliminar))
print("Variables a imputar (<15% NA):", len(vars_imputar))
````

Siguiendo la recomendación metodológica, se eliminan aquellas variables con más de un 15% de datos faltantes, mientras que las variables con menos de un 15% de valores perdidos serán imputadas.

### Revisión de datos faltantes

Se calcula el porcentaje de valores faltantes por variable con el fin de evaluar la calidad de la información disponible.
Siguiendo el criterio indicado en las instrucciones, se imputan aquellas variables con menos de un 15% de valores faltantes y se eliminan las variables que superan dicho umbral, con el objetivo de evitar sesgos y pérdida de información relevante.


````python
#Eliminar variables con demasiados NA
df_etapa1 = df_wb_raw.drop(columns=vars_eliminar).copy()

df_etapa1.shape
````

### Imputación de valores faltantes

Para las variables seleccionadas, se utiliza la mediana como método de imputación, ya que es una medida robusta frente a valores extremos y permite conservar la distribución original de los datos.


````python
#Imputación de variables numéricas (<15%)
#Solo numéricas, usando mediana (robusto a outliers)

# Seleccionar columnas numéricas
num_cols = df_etapa1.select_dtypes(include=['float64', 'int64']).columns

# Imputación por mediana
for col in vars_imputar:
    if col in num_cols:
        df_etapa1[col] = df_etapa1[col].fillna(df_etapa1[col].median())

# Verificación
df_etapa1.isna().sum().sum()
````

### Estadísticas descriptivas

Se calculan estadísticas descriptivas básicas (media, mediana, desviación estándar, valores mínimos y máximos) con el objetivo de caracterizar el comportamiento general de las variables y detectar posibles patrones relevantes.


````python
#Estadisticas descriptivas
desc = df_etapa1[num_cols].describe().T
desc['median'] = df_etapa1[num_cols].median()

desc[['mean', 'median', 'std', 'min', 'max']].round(2)
````

## Etapa 2: Reducción de Dimensionalidad con PCA

El dataset del Banco Mundial contiene una gran cantidad de variables macroeconómicas, demográficas y sociales, muchas de las cuales están altamente correlacionadas entre sí.

El objetivo de esta etapa es aplicar Análisis de Componentes Principales (PCA) con el fin de reducir la dimensionalidad del conjunto de datos, capturar los principales patrones subyacentes y generar un conjunto más compacto de variables que será utilizado posteriormente en los modelos de clasificación del nivel de PIB.

Para ello, se seleccionan únicamente variables numéricas, se estandarizan previamente y se analiza la varianza explicada por cada componente principal, eligiendo un número de componentes que permita explicar entre un 70% y un 90% de la varianza total.


### Selección de variables numéricas

Dado que el PCA solo puede aplicarse a variables numéricas, se seleccionan exclusivamente aquellas variables cuantitativas del dataset.
La variable objetivo (categoría de PIB) se excluye de este proceso, ya que el PCA se utiliza únicamente como técnica de reducción de dimensionalidad para las variables explicativas.


````python
import numpy as np
import pandas as pd

# Seleccionar únicamente variables numéricas
# (usando la base limpia construida en la Etapa 1)
df_numerico = df_etapa1.select_dtypes(include=[np.number]).copy()

print("Dimensión del dataset numérico:", df_numerico.shape)
df_numerico.head()

````

### Tratamiento de valores faltantes

El conjunto de datos presenta valores faltantes en diversas variables.

Dado que el PCA no admite valores NaN, se realiza un proceso de imputación utilizando la mediana de cada variable, lo cual permite conservar la mayor cantidad posible de información sin introducir sesgos extremos, siendo una medida robusta frente a valores atípicos.


````python
from sklearn.impute import SimpleImputer

# Imputación de valores faltantes usando la mediana
imputer = SimpleImputer(strategy="median")
X_imputed = imputer.fit_transform(df_numerico)

print("NaN después de imputación:", np.isnan(X_imputed).sum())
````

````python
df_etapa1.info()
````

### Estandarización de las variables

Antes de aplicar PCA, es necesario estandarizar las variables numéricas, ya que estas se encuentran en distintas escalas de medición.
La estandarización permite que todas las variables contribuyan de manera equitativa a la construcción de los componentes principales.


````python
from sklearn.preprocessing import StandardScaler

# Estandarización de las variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

print("Datos estandarizados:", X_scaled.shape)
````

### Aplicación del Análisis de Componentes Principales (PCA)

Se aplica PCA sobre las variables estandarizadas con el objetivo de analizar cuánta varianza del conjunto de datos original explica cada componente principal.
Este análisis permite determinar cuántos componentes son necesarios para capturar la mayor parte de la información contenida en los datos.


````python
from sklearn.decomposition import PCA

# PCA sin restringir número de componentes
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Varianza explicada
var_explicada = pca.explained_variance_ratio_
var_acumulada = np.cumsum(var_explicada)

# DataFrame con resultados
varianza_df = pd.DataFrame({
    "Componente": np.arange(1, len(var_explicada) + 1),
    "Varianza_Explicada": var_explicada,
    "Varianza_Acumulada": var_acumulada
})

varianza_df.head(10)
````

### Análisis gráfico de la varianza explicada

A continuación, se presentan los gráficos de varianza explicada por componente (scree plot) y varianza acumulada.

Estos gráficos permiten identificar visualmente el número óptimo de componentes principales que permiten explicar entre un 70% y un 90% de la varianza total del dataset.

````python
import matplotlib.pyplot as plt

# Scree plot
plt.figure(figsize=(8,4))
plt.plot(varianza_df["Componente"], varianza_df["Varianza_Explicada"], marker="o")
plt.title("Scree plot: Varianza explicada por componente")
plt.xlabel("Componente")
plt.ylabel("Varianza explicada")
plt.grid(True)
plt.show()

# Varianza acumulada
plt.figure(figsize=(8,4))
plt.plot(varianza_df["Componente"], varianza_df["Varianza_Acumulada"], marker="o")
plt.axhline(0.70, linestyle="--")
plt.axhline(0.90, linestyle="--")
plt.title("Varianza acumulada por componentes")
plt.xlabel("Componente")
plt.ylabel("Varianza acumulada")
plt.grid(True)
plt.show()

````

### Selección del número de componentes principales

En base al análisis de la varianza acumulada, se selecciona el número mínimo de componentes principales que permite explicar al menos el 70% de la varianza total, sin exceder el 90%.

Este criterio busca un equilibrio entre reducción de dimensionalidad y conservación de información relevante.



````python
# Número mínimo de componentes para alcanzar 70% y 90%
k70 = int(np.argmax(varianza_df["Varianza_Acumulada"].values >= 0.70) + 1)
k90 = int(np.argmax(varianza_df["Varianza_Acumulada"].values >= 0.90) + 1)

print("Componentes mínimos para >=70%:", k70)
print("Componentes mínimos para >=90%:", k90)

# Selección final
n_componentes = k70
print("Número de componentes seleccionados:", n_componentes)

````

Construcción del dataset reducido con PCA


Finalmente, se construye un nuevo conjunto de datos que contiene únicamente las componentes principales seleccionadas, junto con la variable objetivo categorizada del PIB.

Este dataset reducido será utilizado como insumo para los modelos de clasificación en la siguiente etapa del proyecto.

````python
# PCA final con número seleccionado de componentes
pca_final = PCA(n_components=n_componentes)
X_pca_final = pca_final.fit_transform(X_scaled)

# DataFrame final con componentes principales
df_pca = pd.DataFrame(
    X_pca_final,
    columns=[f"PC{i+1}" for i in range(n_componentes)]
)

# Agregar variable objetivo
df_pca["PIB_cat"] = df_wb_raw["NY.GDP.MKTP.PP.KD"].values

df_pca.head()
````
